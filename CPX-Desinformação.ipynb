```python
# Importa a biblioteca do Google API
from googleapiclient.discovery import build
import csv

# Define a chave de API do YouTube
API_KEY = 'AIzaSyCF7hR7cl_7fxcTxYozwMpcUDhVzmJP95o'

# Define o nome do canal que voc√™ deseja pesquisar
query = 'CPX'

# Define a data m√≠nima de publica√ß√£o (agosto de 2022)
min_date = '2022-08-01T00:00:00Z'

# Cria um objeto do YouTube Data API v3
youtube = build('youtube', 'v3', developerKey=API_KEY)

# Faz a pesquisa usando o m√©todo search.list do YouTube Data API v3
search_response = youtube.search().list(
    q=query,
    type='video',
    publishedAfter=min_date,
    order='viewCount',
    maxResults=100,
    part='id,snippet',
).execute()

# Cria uma lista para armazenar as informa√ß√µes dos v√≠deos
videos = []

# Percorre os resultados da pesquisa
for search_result in search_response.get('items', []):
    # Obt√©m informa√ß√µes adicionais sobre o v√≠deo usando o m√©todo videos.list do YouTube Data API v3
    video_response = youtube.videos().list(
        id=search_result['id']['videoId'],
        part='snippet,statistics',
    ).execute()
    
    # Extrai as informa√ß√µes relevantes do resultado da pesquisa e do resultado do v√≠deo
    title = video_response['items'][0]['snippet']['title']
    channel_title = video_response['items'][0]['snippet']['channelTitle']
    view_count = video_response['items'][0]['statistics']['viewCount']
    comment_count = video_response['items'][0]['statistics'].get('commentCount', 0)
    published_date = video_response['items'][0]['snippet']['publishedAt']
    
    # Adiciona as informa√ß√µes do v√≠deo √† lista
    videos.append([title, channel_title, view_count, comment_count, published_date])
    
# Cria e escreve os dados em um arquivo CSV
with open('videos.csv', mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL, lineterminator='\n')
    writer.writerow(['T√≠tulo', 'Canal', 'Acessos', 'Coment√°rios', 'Data de publica√ß√£o'])
    for video in videos:
        writer.writerow(video)

# L√™ e exibe o conte√∫do do arquivo CSV
with open('videos.csv', mode='r', encoding='utf-8') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        print(row)

```
